# Calculating Polyspectra from Measurements

The calculation of polyspectra from a measurement is performed using the SignalSnap library. A detail explanation of the functionality of the library can be found on its [GitHub page](https://github.com/MarkusSifft/SignalSnap).

```python
import signalsnap as snp
```

Calculation of the polyspectra

```python
path = 'example_data/long_measurement.h5'
group_key = 'day1'
data_set = 'measurement1'

config = snp.SpectrumConfig(
    path=path,
    group_key=group_key,
    dataset=data_set,
    f_unit='Hz',
    spectrum_size=100,
    order_in='all',
    f_max=0.4e3,
    backend='cpu'
)

spec = snp.SpectrumCalculator(config)

f, s, serr = spec.calc_spec()
```

Plotting of the polyspectra

```python
plot_config = snp.PlotConfig(plot_orders=[2,3,4], sigma=3)
fig = spec.plot(plot_config)
```
